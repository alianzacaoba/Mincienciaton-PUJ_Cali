```python
import pandas as pd
import os.path
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


from pycaret.classification import *
```

```python
path = "../data/"

if not os.path.isfile(path+"seroprevalencia_cali.csv"):
    seroprev_cali_excel = pd.read_excel(path+"datos_seroprevalencia.xlsx",sheet_name='CALI')
    seroprev_bquilla_excel = pd.read_excel(path+"datos_seroprevalencia.xlsx",sheet_name='BQUILLA')
    seroprev_comb_excel = pd.read_excel(path+"datos_seroprevalencia.xlsx",sheet_name='C+B')
    
    seroprev_cali_excel.to_csv(path+"seroprevalencia_cali.csv",index=False)
    seroprev_bquilla_excel.to_csv(path+"seroprevalencia_bquilla.csv",index=False)
    seroprev_comb_excel.to_csv(path+"seroprevalencia_comb.csv",index=False)
    
df_cali    = pd.read_csv(path+"seroprevalencia_cali.csv")
df_bquilla = pd.read_csv(path+"seroprevalencia_bquilla.csv")
df_comb    = pd.read_csv(path+"seroprevalencia_comb.csv")
```

```python
print(df_cali.shape)
print(df_bquilla.shape)
print(df_comb.shape)
```

## DATOS CALI

### 1. Matriz de correlaci√≥n

```python
#get correlations of each features in dataset

corrmat = df_cali.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(20,20))
#plot heat map
g=sns.heatmap(df_cali[top_corr_features].corr(),annot=True,cmap="RdYlGn")
```

```python
## sample returns a random sample from an axis of the object. That would be 950 samples, not 1000
data_cali = df_cali.sample(frac=0.95, random_state=123)
```

```python
# we remove from the original dataset this random data
data_unseen_cali = df_cali.drop(data_cali.index)
```

```python
## we reset the index of both datasets
data_cali.reset_index(inplace=True, drop=True)
data_unseen_cali.reset_index(inplace=True, drop=True)
print('Data for Modeling: ' + str(data_cali.shape))
print('Unseen Data For Predictions: ' + str(data_unseen_cali.shape))
```

```python
model_setup_cali = setup(data=data_cali, target='RESULTADO', session_id=123)
```

```python
best_model_cali = compare_models()
```

```python
print(best_model_cali)
```

#### Modelo CatBoost para Cali

```python
catboost_cali = create_model('catboost')
print(catboost_cali)
```

#### Modelo Random Forest para Cali

```python
rf_cali = create_model('rf')
```

```python
#trained model object is stored in the variable 'dt'. 
print(rf_cali)
```

#### Linear Discriminant Analysis para Cali

```python
lda_cali = create_model('lda')
```

```python
print(lda_cali)
```

#### Tuneo de los modelos de Cali

```python
tuned_catboost_cali = tune_model(catboost_cali)
#tuned model object is stored in the variable 'tuned_catboost_cali'. 
print(tuned_catboost_cali)
```

```python
tuned_rf_cali = tune_model(rf_cali)
#tuned model object is stored in the variable 'tuned_rf_cali'. 
print(tuned_rf_cali)
```

```python
tuned_lda_cali = tune_model(lda_cali)
#tuned model object is stored in the variable 'tuned_lda_cali'. 
print(tuned_lda_cali)
```

#### Curvas ROC de los modelos de cali

```python
## AUC Plot

plot_model(tuned_catboost_cali, plot = 'auc')
```

```python
## AUC Plot

plot_model(tuned_rf_cali, plot = 'auc')
```

```python
## AUC Plot

plot_model(tuned_lda_cali, plot = 'auc')
```

#### Atributos mas importantes de los modelos de Cali

```python
## feature importance

plot_model(tuned_catboost_cali, plot='feature')
```

```python
## feature importance

plot_model(tuned_rf_cali, plot='feature')
```

```python
## feature importance

plot_model(tuned_lda_cali, plot='feature')
```

#### Confusion matrix de los modelos

```python
## Consufion matrix

plot_model(tuned_catboost_cali, plot = 'confusion_matrix')
```

```python
## Consufion matrix

plot_model(tuned_rf_cali, plot = 'confusion_matrix')
```

```python
## Consufion matrix

plot_model(tuned_lda_cali, plot = 'confusion_matrix')
```

#### Terminando los modelos de cali

```python
final_catboost_cali = finalize_model(tuned_catboost_cali)
print(final_catboost_cali)
```

```python
predict_model(final_catboost_cali)

unseen_predictions_catboost_cali = predict_model(final_catboost_cali, data=data_unseen_cali)
unseen_predictions_catboost_cali.head()
```

```python
final_rf_cali = finalize_model(tuned_rf_cali)
print(final_rf_cali)
```

```python
predict_model(final_rf_cali)

unseen_predictions_rf_cali = predict_model(final_rf_cali, data=data_unseen_cali)
unseen_predictions_rf_cali.head()
```

```python
final_lda_cali = finalize_model(tuned_lda_cali)
print(final_lda_cali)
```

```python
predict_model(final_lda_cali)

unseen_predictions_lda_cali = predict_model(final_lda_cali, data=data_unseen_cali)
unseen_predictions_lda_cali.head()
```
