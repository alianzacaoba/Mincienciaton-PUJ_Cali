#### Create a map filling the city boundary with h3 hexagons


```python
resolution = 8
# Get the city boundary and fill it with hexagons
city_boundary_link = "https://raw.githubusercontent.com/finkej/santiago_de_cali/master/mapa_cali/limite_municipal/limite_municipal.json"
city_boundary = gpd.read_file(city_boundary_link)

city_json = json.loads(city_boundary.to_json())['features'][0]['geometry']

list_hexagons = list(h3.polyfill(city_json,
                                 res=resolution, geo_json_conformant=True))

print("At resolution {}".format(resolution),
      "the zone is filled with ", len(list_hexagons), "hexagons")


def reverse_lat_lon(hex_coords):
    return [[coord[1], coord[0]] for coord in hex_coords]


city_hex = gpd.GeoDataFrame({"hex_id": list_hexagons})
city_hex['geometry'] = city_hex.hex_id.apply(lambda x: Polygon(
    reverse_lat_lon(h3.h3_to_geo_boundary(x, geo_json=False))))
```

```python
def map_hex_comuna(df,city_hex, comunas):
    ''' Esta función devuelve una columna nueva en df que será el id de la comuna para cada uno de los hexagonos.
        se puede generar directamente desde lat y lng de los datos en df pero la idea es que se generen desde los hexágonos
        para poder hacer inversas sin problema.
        
        Es una función general, revisar los poligonos de city_hex que estan dentro de columnas, estos dos df pueden ser cualquiera'''
    city_hex2 = city_hex.assign(comuna_id = 0)
    for idx, row in city_hex2.iterrows():
        for idx2, row2 in comunas.iterrows():
            if row2.geometry.contains(row.geometry.centroid):
                city_hex2.at[idx,'comuna_id'] = row2.comuna
    df = df.assign(comuna_id = 0)
    for idx,row in df.iterrows():
        comu = city_hex2[city_hex.hex_id == row.hex_id].comuna_id
        df.at[idx,'comuna_id'] = comu
    return(df, city_hex2)
```

#### funciones de conteo de casos

```python
def hex_to_coordinates(x):
    return Polygon(h3.h3_to_geo_boundary(x,geo_json=True))

def id_to_comuna(x):
    return(comunas[comunas.comuna==22].geometry.iloc[0])

def number_cases_df(df,column):
    data = df.groupby([column])['id'].count().reset_index(name='cases')
    if column == 'hex_id':
        data["geometry"] = data['hex_id'].apply(hex_to_coordinates)
    elif column == 'comuna_id':

        data["geometry"] = data['comuna_id'].apply(id_to_comuna)
    else:
        print('no valid column selected')
    data = gpd.GeoDataFrame(data)
    return data
```

```python
def error_metrics(y_total, preds):
    ''' Función que permite calcular las métricas de error: mse, rmse, mae,mape'''
    mse = mean_squared_error(y_total, preds)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_total, preds)
    mape = mean_absolute_percentage_error(y_total, preds)
    print("MSE: %f" % (mse))
    print("RMSE: %f" % (rmse))
    print("MAE: %f" % (mae))
    print("MAPE: %f" % (mape))
    print()
    
def error_analysis(df, title, path,scaled = False,ac='semana -1',pr = 'predictions'):
    ''' Función realiza un análisis de los errores en los diferentes cuartiles'''
    
    
    #y_total = data_weeks_no_outlier['semana -1']
    #preds = data_weeks_no_outlier['predictions']
    

    if scaled: actual=0
    else: actual=ac
    y_total = df[actual]
    preds = df[pr]
    error_metrics(y_total, preds)
    
    quantiles= list(y_total.quantile([0 ,.25, .5, .75, 1]))  
    queries = [
               y_total < quantiles[1], 
               (y_total >= quantiles[1]) & (y_total< quantiles[2]),
               (y_total >= quantiles[2]) & (y_total < quantiles[3]),
               y_total >= quantiles[3]
              ]
    
    df_q = [df[queries[0]],df[queries[1]],df[queries[2]],df[queries[3]]]
    for i,dfm in enumerate(df_q):
        print('Quantile {}, between {} and {}'.format(i+1,quantiles[i],quantiles[i+1]))
        error_metrics(dfm[actual], dfm[pr])
    
    fig, ax = plt.subplots(figsize=(6, 4))
    y = y_total
    ax.scatter(y, preds)
    ax.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=2)
    ax.plot([y.min(), y.max()], [y.min()*1.25, y.max()*1.25], 'b--', lw=1)
    ax.plot([y.min(), y.max()], [y.min()*0.75, y.max()*0.75], 'b--', lw=1)
    ax.set_xlabel('Real')
    ax.set_ylabel('Predicted')
    plt.title(title)
    plt.savefig(path, dpi=500)
    plt.show()
    
def mean_absolute_percentage_error(y_true, y_pred): 
    ''' métrica MAPE hecha manualmente para trabajar personalmente con las divisiones por cero'''
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    ape = (y_true - y_pred) / y_true
    ape_no_inf = [x if np.abs(x) != float('inf') else 0 for x in ape]
    return np.mean(np.abs(ape_no_inf)) * 100

def plt_map_image(data,col,title,ax):
    '''Función que permite graficar rápidamente los hexagonos de Cali en blanco con algún otro dato'''
    city_hex.plot(color="white", edgecolor='0.1', linewidth=0.4, figsize=(12, 7),ax=ax)
    data.plot(column=col, edgecolor='0.6', linewidth=0.3, cmap='YlOrRd', legend=True, ax=ax);
    ax.set_title(title)
    
### UTILITY FUNCTIONS FOR TARGET SCALING ###
def scale_target(y, mean, std):
    return (y - mean)/std

def reverse_target(pred, mean, std): 
    return pred*std + mean

def plot_results(predicted_data, true_data): 
    fig = plt.figure(facecolor='white') 
    ax = fig.add_subplot(111) 
    ax.plot(true_data, label='True Data') 
    plt.plot(predicted_data, label='Prediction') 
    plt.legend() 
    plt.show()
```

### Long short term model - LSTM modifications and analysis


```python
def split_data_columns(df,n_col,days):
    '''Función que separa los datos de test y training, no uso la función de sklearn debido a que no necesito datos aleatorios,
        si no que debo seguir las series de tiempo.'''
    data_days = df.copy()
    y= data_days.iloc[:,n_col]
    x= data_days.drop(data_days.columns[n_col], axis=1)
    X_test,y_test = x.iloc[:days],y.iloc[:days]
    X_train,y_train = x.iloc[days:],y.iloc[days:]
    X_train = X_train.values.reshape((X_train.shape[0],1, X_train.shape[1]))
    X_test= X_test.values.reshape((X_test.shape[0],1, X_test.shape[1]))
    return X_train, y_train, X_test, y_test
```

```python
def gen_train_data_per_block(df,block,geometry,clinica=None, population=False):
    ''' Función que genera un df con toda la información separada por bloques de días para usar en LSTM fácilmente
    
    geometry: hex_id -> si quiero análisis por hexágonos
              comuna_id -> si quiero análisis por comunas
    '''
    
    ### Se modifica la función para que devuelva la información de los hexágonos pero tambien nuevos 
    ### atributos como IRHF: Infectados totales, recuperados, hospitalizados, fallecidos cada día.
    
    if clinica==None: filtered = df.copy()
    else: filtered = df[df['nom_upgd']==clinica]
        
    if population: change_name = 'normalized_cases'
    else: change_name = 'cases'
    
    df_outer = city_hex.copy()
    if geometry == 'hex_id':
        df_outer.drop(columns=['comuna_id'],inplace=True)
        
    elif geometry == 'comuna_id':
        df_outer.drop(columns=['hex_id'],inplace=True)
        df_outer = df_outer.drop_duplicates(subset='comuna_id', keep='first', inplace=False).reset_index(drop=True)
        
    df_dates = df.sort_values(by='date',ascending=False)
    df_dates = df_dates[df_dates.date < '2020-07-01']
    df_per_block = []
    
    for idx, day in filtered.groupby(pd.Grouper(key="date", freq=block)):
        df_per_block.append(number_cases_df(pd.DataFrame(day),geometry))
        if population:
            data = df_per_block[-1].iloc[:,:-1]
            if geometry == 'hex_id':
                data = pd.merge(left=data, right=population_per_hex, on='hex_id')
            elif geometry == 'comuna_id':
                data = pd.merge(left=data, right=population_per_commune, on='comuna_id')
            data['normalized_cases'] = data['cases'] / data['population']
            data.drop(columns=['cases','population'],inplace=True)
            #data = df_outer.rename(columns={'normalized_cases':'cases'})
            df_per_block[-1] = gpd.GeoDataFrame(data)     
            
        df_per_block[-1]=df_per_block[-1].rename(columns={change_name: idx})
        
    for i,daf in enumerate(df_per_block):        

        df_outer = pd.merge(df_outer, daf, on=geometry, how='outer')
        df_outer['geometry'] = df_outer.apply(lambda row: row['geometry_y'] if row['geometry_y']!=None else row['geometry_x'], axis=1)
        df_outer.drop(columns=['geometry_x', 'geometry_y'],inplace=True)
        #df_outer = df_outer.rename(columns={change_name: 'semana {0}'.format(i)})

    df_outer.iloc[:,1:-1] = df_outer.iloc[:,1:-1].fillna(0)
    df_outer = df_outer.T
    df_outer.columns = df_outer.iloc[0].values
    df_outer = df_outer.drop([geometry,'geometry'])

    df_outer = df_outer.sort_index(ascending=False)

    
    ## código que agrega irhf
    df_outer = df_outer.assign(infectados = 0 , 
                               recuperados = 0 , 
                               hospitalizados = 0, 
                               fallecidos = 0, 
                               eve_345 = 0, 
                               eve_346 = 0,
                               eve_348 = 0, 
                               notificaciones = 0, 
                               fec_con = 0)
    
    lista_infect = pd.to_datetime(filtered.date)
    lista_muertes = pd.to_datetime(filtered.fec_def_[pd.notna(filtered.fec_def_)])
    lista_rec =  pd.to_datetime(filtered.fec_hos_[(pd.notna(filtered.fec_hos_)) & (filtered.vive == 1)])
    lista_hosp =  pd.to_datetime(filtered.fec_hos_[pd.notna(filtered.fec_hos_)])
    
    ## Excepciones
    filtered.fec_not = filtered.fec_not.replace({'0020-06-25':'2020-06-25', '0020-06-26':'2020-06-26'})
    
    lista_not =  pd.to_datetime(filtered.fec_not[pd.notna(filtered.fec_not)])
    lista_con =  pd.to_datetime(filtered.fec_con_[pd.notna(filtered.fec_con_)])
    
    ## cuenta de los diferentes códigos de eventos.
    temp = filtered[['cod_eve','date']]
    temp = pd.get_dummies(temp,prefix=['eve'], columns = ['cod_eve'], drop_first=False)
    
    for idx, day in df_outer.iterrows():        
        df_outer.at[idx,'infectados']     = lista_infect[lista_infect == idx ].shape[0]
        df_outer.at[idx,'fallecidos']     = lista_muertes[lista_muertes == idx ].shape[0]
        df_outer.at[idx,'recuperados']    = lista_rec[lista_rec == idx ].shape[0]
        df_outer.at[idx,'hospitalizados'] = lista_hosp[lista_hosp == idx ].shape[0]
        
        df_outer.at[idx,'notificaciones'] = lista_not[lista_not == idx ].shape[0]
        df_outer.at[idx,'fec_con']        = lista_con[lista_con == idx ].shape[0]
        
        df_outer.at[idx,'eve_345']        = temp[(temp.date == idx)& (temp['eve_345.0']==1) ].shape[0]
        df_outer.at[idx,'eve_346']        = temp[(temp.date == idx)& (temp['eve_346.0']==1) ].shape[0]
        df_outer.at[idx,'eve_348']        = temp[(temp.date == idx)& (temp['eve_348.0']==1) ].shape[0]
        
    return df_outer
```

```python
data_days = gen_train_data_per_block(df,'1D', geometry='comuna_id',population = True)
data_days = data_days[(data_days.index > datetime.date(2020, 3, 1)) & (data_days.index < datetime.date(2020, 6, 30))]
```

```python
## Nota: se pueden mejorar la siguientes dos funciones convirtiendolas en una, pero queda mas complicado al entender.
##       se propone usar herencia si se utilizarán en el código final.

def lstm_feature_importance(data_days,n_test,name_error,path_error,name_map,path_map,geometry):
    '''Función que realiza las predicciones con los datos completos, entrega resultado de predicciones,
       también devuelve los mejores hexágonos para la prediccion en cada sector.'''
    ## data_days.shape[1]-9  /// el 9 viene siendo las 9 nuevas columnas
    dict_pred = {}
    for n_col in range(data_days.shape[1]-9):
        X_train, y_train, X_test, y_test = split_data_columns(data_days,n_col,n_test)

        ### SCALE TARGET ###

        mean_train = y_train.mean()
        std_train = y_train.std()
        
        
        y_train_seq = scale_target(y_train, mean_train, std_train)
        y_test_seq = scale_target(y_test, mean_train, std_train)


        scaler = StandardScaler()
        X_train = scaler.fit_transform(X_train.reshape(-1,len(X_train[0][0]))).reshape(X_train.shape).astype('float32')
        X_test = scaler.transform(X_test.reshape(-1,len(X_train[0][0]))).reshape(X_test.shape).astype('float32')
        print(y_train_seq.shape,  y_test_seq.shape)

        model = Sequential()
        model.add(LSTM(100 ,input_shape=(X_train.shape[1], X_train.shape[2])))
        model.add(Dropout(0.2)) 
        model.add(Dense(1))
        model.compile(loss='mean_squared_error', optimizer='adam')

        print ('model compiled {}'.format(n_col))

        print (model.summary())

        es = EarlyStopping(patience=30, verbose=1, min_delta=0.001, monitor='val_loss', mode='auto')  
        model.fit(X_train, y_train_seq, batch_size=4, epochs=50, validation_data = (X_test, y_test),callbacks=[es]) ## REVISAR ESTE VALIDATION DATA


        ### MAKE PREDICTION ON TEST ###

        predicted_cases = reverse_target(model.predict(X_test).ravel(), mean_train, std_train)


        df_y = pd.DataFrame(y_test)
        df_y = df_y.assign(pred=predicted_cases)

        background = X_test[:]
        explainer = shap.DeepExplainer(model,background)
        shap_values = explainer.shap_values(background)
        names = list(data_days.drop(data_days.columns[n_col], axis=1).columns)
        
        reshaped_shap_values = np.reshape(shap_values[0], (len(background), data_days.shape[1]-1))
        reshaped_background = np.reshape(background, (len(background), data_days.shape[1]-1))

        vals= np.abs(reshaped_shap_values).mean(0)
        feature_importance = pd.DataFrame(list(zip(names, vals)), columns=['col_name','feature_importance_vals'])
        feature_importance.sort_values(by=['feature_importance_vals'], ascending=False,inplace=True)


        dict_pred[data_days.columns[n_col]] = [df_y.copy(),feature_importance]
    
        K.clear_session()
    
    
    pred_lstm = city_hex.copy()
    if geometry == 'hex_id':
        pred_lstm.drop(columns=['comuna_id'],inplace=True)
        
    elif geometry == 'comuna_id':
        pred_lstm.drop(columns=['hex_id'],inplace=True)
        pred_lstm = pred_lstm.drop_duplicates(subset='comuna_id', keep='first', inplace=False).reset_index(drop=True)
        
    pred_lstm = pred_lstm.assign(real = float(1), pred = float(1))

    for index, row in pred_lstm.iterrows():
        sum_real = dict_pred[row[geometry]][0][row[geometry]].sum()
        sum_pred = dict_pred[row[geometry]][0]['pred'].sum()

        pred_lstm.at[index,'real'] = sum_real
        pred_lstm.at[index,'pred'] = sum_pred

    if geometry == 'hex_id':
        pred_lstm = pred_lstm[pred_lstm['hex_id'] != '8866f1c427fffff']
        data = pd.merge(left=pred_lstm, right=population_per_hex.iloc[:,:-1], on='hex_id')
    elif geometry == 'comuna_id':
        data = pd.merge(left=pred_lstm, right=population_per_commune.iloc[:,:-1], on=geometry)
        
    pred_lstm_cases = pred_lstm.assign(pred = list(data['pred']*data['population']),real = list(data['real']*data['population']))

    error_analysis(pred_lstm_cases, name_error, path=path_error,scaled = False,ac='real',pr = 'pred')

    
    fig, axs = plt.subplots(1, 2,figsize=(12, 6))
    map_hex =gpd.GeoDataFrame(pd.merge(pred_lstm_cases[['comuna_id','real','pred']],city_hex , on = 'comuna_id', how='left'))
    ##comunas.plot(color="white", edgecolor='0.1', linewidth=0.4, figsize=(12, 7),ax=axs[0])
    ##aaa = gpd.GeoDataFrame(pd.merge(pred_lstm_cases[['comuna_id','real','pred']], comunas[['comuna','geometry']], left_on='comuna_id', right_on='comuna', how='left').drop('comuna', axis=1))
    map_hex.plot(column='real', edgecolor='0.6', linewidth=0.3, cmap='YlOrRd', legend=True, ax=axs[0])
    axs[0].set_title('Real'+name_map)

    #comunas.plot(color="white", edgecolor='0.1', linewidth=0.4, figsize=(12, 7),ax=axs[1])
    map_hex.plot(column='pred', edgecolor='0.6', linewidth=0.3, cmap='YlOrRd', legend=True, ax=axs[1])
    axs[1].set_title('Pred '+name_map)
    plt.savefig(path_map, dpi=500)
    plt.show()
    
    return dict_pred,pred_lstm,pred_lstm_cases

```

```python
import collections

def dict_occurrences_hexagons(dict_pred_sel_hexa,n_best):
    dict_occurrences = dict.fromkeys(dict_pred_sel_hexa.keys(),0)
    dict_occurrences['infectados'] = 0
    dict_occurrences['hospitalizados'] = 0
    dict_occurrences['recuperados'] = 0
    dict_occurrences['fallecidos'] = 0
    dict_occurrences['notificaciones'] = 0
    dict_occurrences['fec_con'] = 0
    dict_occurrences['eve_345'] = 0
    dict_occurrences['eve_346'] = 0
    dict_occurrences['eve_348'] = 0
    for k,v in dict_pred_sel_hexa.items():
        iterator = list(v[1].col_name[:n_best])
        for iter in iterator:
            dict_occurrences[iter]+=1
    
    return sorted(dict_occurrences.items(), key=lambda x: x[1], reverse=True)
```

```python
occurrences_30 = dict_occurrences_hexagons(dict_pred_sel_hexa,30)
occurrences_15 = dict_occurrences_hexagons(dict_pred_sel_hexa,15)
occurrences_5 = dict_occurrences_hexagons(dict_pred_sel_hexa,5)
```
